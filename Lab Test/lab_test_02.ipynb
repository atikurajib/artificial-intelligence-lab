Q1: Write a program to count how many vowels and consonants are in a given string.

```python
given_string = input("Enter a string: ")

vowels = "aeiouAEIOU"

vowel_count = 0

consonant_count = 0
for char in given_string:
    if char.isalpha():
        if char in vowels:
            vowel_count += 1
        else:
            consonant_count += 1

print("Total number of vowels:", vowel_count)
print("Total number of consonants:", consonant_count)
```

Q2: Create a function called find_grade(score) that returns:
- A for score ≥ 80
- B for 70–79
- C for 60–69
- F for < 60

Take marks of 10 students, store in a dictionary (name: grade), and print the dictionary.

```python
def find_grade(score):
    if score >= 80:
        return 'A'
    elif 70 <= score < 80:
        return 'B'
    elif 60 <= score < 70:
        return 'C'
    else:
        return 'F'

students_scores = {

    "Amir": 85,
    "Reza": 72,
    "Sabina": 65,
    "Tania": 58,
    "Nabil": 90,
    "Tomal": 77,
    "Rafi": 61,
    "Sadia": 49,
    "Jamil": 83,
    "Munia": 69

}

all_students_grades = {}

for name, score in students_scores.items():
    all_students_grades[name] = find_grade(score)
print(all_students_grades)
```
